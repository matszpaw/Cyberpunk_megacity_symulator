```python
def kaczka(n):
    for x in range(n):
        print(x*4)
```

```python
kaczka(100)
```

```python
import numpy as np

def iterator(world):
    shape = np.shape(world)
    for row in range(shape[0]):
        for column in range(shape[1]):
            return world[row,column]
```

```python
iterator(np.array(([3,4],[2,6]))
```
